```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import random

%matplotlib inline
```

```python
IMAGE_PATH = "boats.jpg"

img = cv2.imread(IMAGE_PATH)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### 1. Crie primeiramente um script ou Jupyter notebook que receba uma imagem e a suavize com filtros implementados. Tente implementar filtros da própria biblioteca OpenCV, mas utilizando parâmetros alternáveis como tamanho do kernel e sigma, caso seja o gaussiano, por exemplo.

```python
# Sigma
sigma = 20
# Kernel
kernel = 5

blur = cv2.GaussianBlur(img, (kernel, kernel), sigma)

# Visualização das imagens
plt.imshow(img, cmap="gray"),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.show()
plt.imshow(blur, cmap="gray"),plt.title('Imagem filtrada')
plt.xticks([]),plt.yticks([])
plt.show()
```

### 2. Na mesma imagem ou em uma outra, aplique alguns filtros para enaltecer as bordas, linhas e contornos.

```python
# Filtro de Laplacian
kernel = np.array((
  [0, 1, 0],
  [1, -4, 1],
  [0, 1, 0]), dtype="int")
# Realiza a convolução
filtered = cv2.filter2D(img, -1, kernel)
# Visualização das imagens
plt.imshow(img, cmap="gray"),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.show()
plt.imshow(filtered, cmap="gray"),plt.title('Imagem filtrada')
plt.xticks([]),plt.yticks([])
plt.show()
```

### 4. Como foi mencionado, a utilização de filtros é uma febre nas redes sociais. De olho na segurança da informação, elabore um texto identificando, pelo menos, 3 riscos envolvidos no ato do usuário aplicar filtros em fotos no perfil, bem como, explicando como ele poderia se proteger desses riscos quando utiliza filtros.

Um estudo de 2021 realizado por pesquisadores da City University of London explorou os efeitos negativos dos filtros sobre a saúde mental. Entre os participantes estavam 175 jovens mulheres e pessoas não-binárias, entre 18 e 30 anos, sendo a maioria de 20 e 21 anos. Uma das descobertas mais marcantes foi que 90% das jovens mulheres do estudo relataram usar filtros ou editar suas fotos.

O estudo também encontrou respostas para a pergunta: Qual é o filtro que todos usam mais? Os cinco filtros ou edições mais comuns foram aqueles usados para uniformizar o tom de pele, clarear a pele, branquear os dentes, bronzear a pele e tirar o peso. Além disso, os participantes usavam filtros nas mídias sociais para remodelar a mandíbula ou o nariz, fazer com que seus lábios parecessem mais cheios, e fazer com que seus olhos parecessem maiores.

> "As jovens nos disseram que se sentem sob uma pressão considerável para se apresentarem como divertidas, felizes e sociáveis, bem como sem esforço - refletindo as formas como as pressões da aparência se estenderam para apresentar 'um eu perfeito'. Elas refletiram com autocrítica sobre todos os aspectos de si mesmas, incluindo suas roupas, sapatos, estilo e aparência, e muitas vezes preocupadas com a possibilidade de "derrubar amigos" em uma foto de grupo. É difícil exagerar o sentido palpável da pressão que eles transmitiam".

> > A autora do estudo Rosalind Gill, professora de Análise Social e Cultural
> > City University of London


#### Efeitos negativos dos filtros
Analisando a associação entre filtros e saúde mental, o estudo encontrou as seguintes estatísticas sobre os efeitos negativos dos filtros:

94% dos participantes disseram que se sentiram pressionados a olhar para um determinado lado, e mais da metade deles disseram que a pressão era intensa. 
70 por cento sentiu pressão para mostrar uma "vida perfeita".
86% disseram que suas representações nas mídias sociais não refletiam sua vida real.
Os pesquisadores declararam: "As jovens mulheres nos disseram que estes padrões [de atratividade] são 'inatingíveis' e 'tóxicos' e que os estão desgastando e tornando-os infelizes". Além disso, os jovens com baixa auto-estima e má imagem corporal são mais propensos a usar filtros, reforçando ainda mais a crença negativa de que sua aparência não é suficientemente boa.

Além disso, estes sentimentos e comportamentos começam cedo. Pesquisas conduzidas pelo Projeto Dove de Auto-Estima em 2020 descobriram que 80% das meninas já baixaram um filtro ou usaram um aplicativo para mudar a aparência nas fotos até os 13 anos de idade.

#### Dismorfia Corporal e Filtros de Mídias Sociais 
Outra ligação entre os filtros e a saúde mental é o impacto negativo na imagem corporal. Pesquisas mostram que os jovens adultos que freqüentemente usam filtros nas mídias sociais frequentemente têm sentimentos crescentes de insatisfação com seu rosto e corpo reais. Eles não só estão comparando sua aparência com imagens "perfeitas" de celebridades e pares, mas também estão se julgando contra seus próprios selos filtrados. Esta comparação constante pode causar estragos na imagem e na auto-estima do corpo. 

Na verdade, a síndrome da colisão da comparação, a dismorfia corporal e as redes sociais criaram uma "dismorfia filtrante", também chamada de dismorfia de Snapchat. A dismorfia, também chamada de distúrbio dismórfico corporal, é uma condição de saúde mental na qual a pessoa está obcecada com as falhas percebidas em seu rosto ou corpo. A comparação das mídias sociais e os filtros nas mídias sociais fazem com que as pessoas se esforcem por padrões de beleza irreais, contribuindo para o aumento da dismorfia.

Assim, os jovens com dismorfia filtrada experimentam uma sensação de desconexão entre o seu aspecto real e as imagens editadas que compartilham com o mundo - uma espécie de auto-objectificação. Pode ser por isso que as pessoas que usam aplicativos de edição de fotos regularmente também são mais propensas a considerar a possibilidade de fazer cirurgia estética. Outro aspecto desta questão é a forma como os filtros são intrinsecamente racistas - iluminando a pele, estreitando o nariz e mudando as características faciais para um padrão de beleza eurocêntrico.
